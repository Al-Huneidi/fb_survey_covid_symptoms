```python
# https://covidmap.umd.edu/api.html
# COVID-19 World Survey Data API
```

```python
import requests
import json
import pandas as pd
```

```python
# request New Zealand data from api
response = requests.get("https://covidmap.umd.edu/api/resources?indicator=covid&type=smoothed&country=New Zealand&daterange=20200401-20200430").text
```

```python
#convert json data to dic data for use!
jsonData = json.loads(response)
jsonData
```

```python
# reference : https://www.geeksforgeeks.org/convert-json-to-dictionary-in-python/
print(jsonData['data'][0])
```

```python
NZLdata=jsonData['data']
NZLdata
```

```python
NZL_df=pd.DataFrame.from_dict(NZLdata)
NZL_df.head()
```

```python
# request Brazil data from api
response_bra = requests.get("https://covidmap.umd.edu/api/resources?indicator=covid&type=smoothed&country=Brazil&daterange=20200401-20200430").text
```

```python
#convert json data to dic data for use!
jsonData_bra = json.loads(response_bra)
jsonData_bra
```

```python
BRAdata=jsonData_bra['data']
BRAdata
```

```python
BRA_df=pd.DataFrame.from_dict(BRAdata)
BRA_df.head()
```
